```python
df_no_long_tails = df
df_no_long_tails.v1 = v1_logs
df_no_long_tails.amount = amount_logs
```

```python
#computing feature importance
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df_no_long_tails, test_size=0.20, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
len(df), len(df_train), len(df_val), len(df_test)
```

```python
y_train = df_train["class"].values
y_val = df_val["class"].values
y_test = df_test["class"].values
```

```python
del df_train["class"]
del df_val["class"]
del df_test["class"]
```

```python
df_train.reset_index(drop=True, inplace=True)
df_val.reset_index(drop=True, inplace=True)
df_test.reset_index(drop=True, inplace=True)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

```python
def plot_features_and_get_auc(feature_names):
    results_dict = {}
    plt.figure(figsize=(5,5))
    plt.plot ([0,1], [0,1], label = 'Random', linestyle = '--', alpha=0.5)
    for col in feature_names:
        fpr, tpr, thresholds = roc_curve(y_true=y_train, y_score=df_train[col])
        auc_score_result = auc(fpr, tpr)
        if (auc_score_result < 0.5):
            fpr, tpr, thresholds = roc_curve(y_true=y_train, y_score=-df_train[col])        
            auc_score_result = auc(fpr,tpr)
        plt.plot(fpr,tpr, label =col)        
        results_dict[col]=auc_score_result
    
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.legend()

    results_df = pd.DataFrame.from_dict(results_dict, orient='index', columns=['auc']).reset_index()
    results_df.columns = ['feature', 'auc']  
    return results_df
```

```python
cols_of_interest = [f'v{i}' for i in range(1, 14)]
df_aucs_v1_to_14=plot_features_and_get_auc(cols_of_interest)
```

```python
df_aucs_v1_to_14.sort_values(by='auc')
```

```python
#v13 is a low importance feature in the previous subset
```

```python
#analizing importance of features v15 to v29
```

```python
cols_of_interest = [f'v{i}' for i in range(15, 28)]
df_aucs_v15_to_28 = plot_features_and_get_auc(cols_of_interest)
```

```python
df_aucs_v15_to_28.sort_values(by='auc')
```

```python
#feature v15 and features v22 to v26 are low imporatance features
```

```python
#analizing importance of time and amount
```

```python
cols_of_interest = ['time','amount']
df_aucs_time_and_amount = plot_features_and_get_auc(cols_of_interest)
```

```python
df_aucs_time_and_amount.sort_values(by="auc")
```

```python
#amount and time are low importance features
```
