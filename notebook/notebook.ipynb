```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

```python
def plot_features_and_get_auc_for_features(feature_names):
    results_dict = {}
    plt.figure(figsize=(5,5))
    plt.plot ([0,1], [0,1], label = 'Random', linestyle = '--', alpha=0.5)
    for col in feature_names:
        fpr, tpr, thresholds = roc_curve(y_true=y_train, y_score=df_train[col])
        auc_score_result = auc(fpr, tpr)
        if (auc_score_result < 0.5):
            fpr, tpr, thresholds = roc_curve(y_true=y_train, y_score=-df_train[col])        
            auc_score_result = auc(fpr,tpr)
        plt.plot(fpr,tpr, label =col)        
        results_dict[col]=auc_score_result
    
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.legend()

    results_df = pd.DataFrame.from_dict(results_dict, orient='index', columns=['auc']).reset_index()
    results_df.columns = ['feature', 'auc']  
    return results_df
```

```python
cols_of_interest = [f'v{i}' for i in range(1, 14)]
df_aucs_v1_to_14=plot_features_and_get_auc(cols_of_interest)
```

```python
# Training a logistic regression without applying logarithm to amount and v1
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
def find_logistic_regression_accuracy(df_train, y_train, df_val, y_val, threshold=0.5):
    dv= DictVectorizer(sparse=False)
    model = LogisticRegression(solver="liblinear", max_iter=200, random_state=1)
    
    _train_dicts = df_train.to_dict(orient='records')
    _X_train = dv.fit_transform(_train_dicts)
    model.fit(_X_train, y_train)
    
    _val_dicts = df_val.to_dict(orient='records')    
    _X_val = dv.transform(_val_dicts)
    y_pred_proba = model.predict_proba(_X_val)[:,1]
    decision = (y_pred_proba>=threshold)
    
    accuracy = (y_val == decision).mean().round(5)
    return accuracy, model, X_train, X_val
```

```python
accuracy_without_logs, model, X_train, X_val = find_logistic_regression_accuracy(df_train, y_train, df_val, y_val)
accuracy_without_logs
```

```python
# the model has a high accuracy with a threshold of 0.5 and without applying the logarithm to amount and v1
```

```python
# Looking at the coefficients
```

```python
weights = model.coef_[0]
weights_df = pd.DataFrame({'Feature': dv.get_feature_names_out(), 'Weight': weights, 'Abs_Weight': abs(weights)})
weights_df = weights_df.sort_values(by='Abs_Weight', ascending=False)
weights_df
```

```python
#Here the amount and time has a low weight, which aligns with to the low correlation between these features and the target variable. 
# the low weight for v12 is a surprise, given that it is a high importance feature as per previous analysis
```
